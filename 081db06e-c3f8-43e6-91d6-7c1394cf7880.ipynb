```python
# Calcule a quantidade de minutos gastos por cada usuário por mês. Salve o resultado.

#Primeiro vamos agrupar os usuários e agrupá-los
minutes_per_month = calls.groupby(['user_id', 'month'])['duration'].sum().reset_index()

#Para finalizar, jogamos os dados filtrados criando um outro DatFrame
minutes_per_month.to_csv('minutes_per_month.csv', index=False)
print(minutes_per_month)
print(minutes_per_month.info())
```

```python
# Calcule o número de mensagens enviadas por cada usuário por mês. Salve o resultado.

#Primeiro, vamos precisar converter o tipo de dados na coluna de data e logo depois extratir o mês.
messages['message_date'] = pd.to_datetime(messages['message_date'])
messages['month'] = messages['message_date'].dt.month

#Agora vamos ordenar os dados filtrados.
messages_per_month = messages.groupby(['user_id','month']).size().reset_index(name='num_messages')

#Criando um DataFrame separados para os dados.
messages_per_month.to_csv('messages_per_month.csv', index=False)
print(messages_per_month.info())
```

```python
# Calcule o volume de tráfego de internet usado por cada usuário por mês. Salve o resultado.

#Assim como fizemos com os outros dados, vamos converter o formato de horas no DatFrame
internet['session_date'] = pd.to_datetime(internet['session_date'])
internet['month'] = internet['session_date'].dt.month

#Para efetuar esse cálculo, vamos agrupar os valores pelo ID e somá-los, e assim depois, criar um novo DataFrame. 
monthly_traffic = internet.groupby(['user_id','month'])['mb_used'].sum().reset_index()

monthly_traffic.to_csv('monthly_traffic.csv', index=False)
print(monthly_traffic.head())
print()
print(monthly_traffic.info())
```

[Calcule a receita mensal para cada usuário: subtraia o limite gratuito do pacote do número total de chamadas, mensagens de texto e dados; multiplique o resultado pelo valor do plano de chamadas; adicione o preço mensal dependendo do plano. Importante: isso pode não ser tão trivial quanto apenas algumas linhas, dadas as condições do plano. Portanto, não há problema em gastar algum tempo nisso.]

```python
# Calcule a receita mensal para cada usuário

#O primeiro passo foi definir uma função para efetuar o cálculo
def calculate_monthly_revenue(row):
    plan = row['plan']


#Aqui vamos puxar dos DataFrames os valores totais de cada item.    
    total_calls = row['call_num'] 
    total_messages = row['num_messages']
    total_data = row['mb_used']

#Depois selecionamos apenas os valores que já são inclusos no plano.    
    plan_limits = plans[plans['plan_name'] == plan].iloc[0]
    call_included = plan_limits['minutes_included']
    messages_included = plan_limits['messages_included']
    mb_included = plan_limits['mb_per_month_included']

#Nessa parte aqui estou calculando o total de valor excedente de cada item.    
    extra_call = max(total_calls - call_included, 0)
    extra_messages = max(total_messages - messages_included, 0)
    extra_data = max(total_data - mb_included, 0)


    revenue = 0

#Usando o método IF para efetuar os cálculos de valores excedentes.
    if extra_call > 0:
        revenue += extra_call * plan_limits['usd_per_minute']

    if extra_messages > 0:
        revenue += extra_messages * plan_limits['usd_per_message']

    if extra_data > 0:
#Nesse caso específico eu tive que aplicar uma conversão de dados, já que o plano cobra em GB e o nosso novo DataFrame estava em MB.        
        extra_data_gb = extra_data / 1024
        revenue += extra_data_gb * plan_limits['usd_per_gb']


    revenue += plan_limits['usd_monthly_pay']

    return revenue

#Criando um novo DataFrame aplicando a função para calcular a receita mensal.
merged_data['monthly_revenue'] = merged_data.apply(calculate_monthly_revenue, axis=1)

#Logo após criar o novo DataFrame de receita mensal, precisei efetuar umas mudanças, pois a coluna 'monthly_revenue' possuía muitas casas depoi da vírgula.
merged_data['monthly_revenue'] = merged_data['monthly_revenue'].round(2)

print(merged_data.head())
print()
print(merged_data.info())
print()
merged_data.sample(5)
```

## Teste hipóteses estatísticas

[Teste a hipótese de que as receitas médias dos usuários dos planos de chamadas Ultimate e Surf são diferentes.]

[Formule as hipóteses nula e alternativa, escolha o teste estatístico, escolha o valor alfa.]

```python
# Teste as hipóteses

from scipy.stats import ttest_ind, levene

'''
H0: As receitas médias dos planos Surf e Ultimate são iguais.
H1: As receitas médias dos planos Surf e Ultimate não são iguais.
'''
# Separarando as receitas por plano
revenue_surf = merged_data[merged_data['plan'] == 'surf']['monthly_revenue']
revenue_ultimate = merged_data[merged_data['plan'] == 'ultimate']['monthly_revenue']

# T-testes de amostras independentes
stat, p_levene = levene(revenue_surf, revenue_ultimate)

equal_var = p_levene > 0.05

t_statistic, p_value = ttest_ind(revenue_surf, revenue_ultimate, equal_var=equal_var)

alpha = 0.05

print(f"Teste de Levene p-value: {p_levene}")
print(f"Teste t p-value: {p_value}")





# Aqui, por fim, verificamos se rejeitamos ou não a hipótese nula
if p_value < alpha:
    print("Rejeitamos a hipótese nula. As receitas médias dos planos Surf e Ultimate são estatisticamente diferentes.")
else:
    print("Não rejeitamos a hipótese nula. Não há evidências suficientes para afirmar que as receitas médias dos planos Surf e Ultimate são diferentes.")
```
